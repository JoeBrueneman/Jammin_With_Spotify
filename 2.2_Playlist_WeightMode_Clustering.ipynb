# Clustering Playlists

## Processing Data

```python
# Import libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
# Read the csv file into a pandas DataFrame
playlist_mean_mode_df = pd.read_csv(Path("../Resources/playlist_weight_mode.csv"))

# Review the DataFrame
playlist_mean_mode_df
```

```python
playlist_mean_mode_df.columns
```

```python
# Retain the useful columns to fit the model
playlist_mean_mode_df = playlist_mean_mode_df[['pid', 'weight_danceability', 'weight_energy',
       'mode_key', 'proportion_key_none', 'proportion_key_0',
       'proportion_key_1', 'proportion_key_2', 'proportion_key_3',
       'proportion_key_4', 'proportion_key_5', 'proportion_key_6',
       'proportion_key_7', 'proportion_key_8', 'proportion_key_9',
       'proportion_key_10', 'proportion_key_11', 'weight_loudness',
       'mode_mode', 'proportion_mode_minor', 'proportion_mode_major',
       'weight_speechiness', 'weight_acousticness', 'weight_instrumentalness',
       'weight_liveness', 'weight_valence', 'weight_tempo',
       'weight_duration_ms']]
# Check out `playlist_mean_df`
playlist_mean_mode_df
```

## Scaling

```python
# Import libraries and dependencies
from sklearn.preprocessing import StandardScaler
```

```python
# Creating `scaler` 
scaler = StandardScaler()
# Scale the `playlist_mean_df` and drop the 'pid' column
scaler.fit(playlist_mean_mode_df.drop('pid',axis=1))
scaled_playlist_mean_mode = scaler.transform(playlist_mean_mode_df.drop('pid',axis=1))
```

```python
# Create `scaled_playlist_mean_df` for the scaled_playlist_mean
scaled_playlist_mean_mode_df = pd.DataFrame(scaled_playlist_mean_mode,columns=playlist_mean_mode_df.drop('pid',axis=1).columns, index=playlist_mean_mode_df.index)
# Check out `scaled_playlist_mean_df`
scaled_playlist_mean_mode_df
```

## Finding K Value
* Elbow Method vs. Silhouette Score

```python
# Import libraries and dependencies
from sklearn.cluster import KMeans, AgglomerativeClustering, Birch
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import hvplot.pandas
import time
```

```python
# Set up the KMeans model where k is the number for clusters ranging from 1 to 10
start_time = time.time()

inertias =[]
k = list(range(1,10))

for i in k:
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=42)
    kmeans.fit(scaled_playlist_mean_mode_df)
    inertias.append(kmeans.inertia_)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Create elbow_df and sil_score_df
elbow_df = pd.DataFrame({
    "k": k,
    "inertia": inertias
})
# Check out `elbow_df`
elbow_df
```

```python
# Plot elbow curve
elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

## Clustering & Visualization

### Use Silimarity to make best 100 songs
