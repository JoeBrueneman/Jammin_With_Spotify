```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import json
import csv
import time
from IPython.display import clear_output
```

```python
# Read the track_ids as a flat list
with open('tracks_URI.csv', newline='') as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    tracks_list = []
    for _ in reader:
        tracks_list.append(_[0])
```

```python
# Create the session with Spotify
spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())

```

```python
# The first number defines the beginning of the slice of tracks to be queried
# The second number is the length of the existing dataframe after the last timeout/Server_Disconnect
a=1000000 + 298300  

# This is the end of the slice of tracks to be queried
b=1500000
```

```python
# a=0
# b=2262292
for i in range(a,b,100):
    x=tracks_list[i:i+100]
    results = spotify.audio_features(x)
    for j in range(0,100):
        df_dictionary = pd.DataFrame([results[j]])
        tracks_features_df = pd.concat([tracks_features_df, df_dictionary], ignore_index=True)
    print(i)
    clear_output(wait=True)
    time.sleep(0.5)
```
